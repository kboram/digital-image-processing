```python
outimg = gimg1.copy()
#bllurinage = cv2.GaussianBlur(outimg, (7, 7), 5)
blurimg = cv2.medianBlur(outimg, 7)

x = 150
y = 300
size = 30

for h in range(y, y+size):
  for w in range(x, x+size):
    #outimg[h][w] = 255
    outimg[h][w] = blurimg[h][w]

cv2_imshow(outimg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg1=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg1)
```

```python
outimg = cv2.GaussianBlur(gimg1,(7, 7),2)
cv2_imshow(outimg)
```

```python
edgeimage = cv2.addWeighted(gimg1, 1, outimg, -1, 0)
cv2_imshow(edgeimage)
```

```python
unsharpimage = cv2.addWeighted(gimg1, 1, edgeimage, -1, 0)
cv2_imshow(unsharpimage)
```

```python
kernel3 = np.ones((3, 3), np.float32)/9
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = -1/9
kernel3[1][1] = -1/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
```

```python
output = cv2.filter2D(gimg1, -1, kernel3)
cv2_imshow(output)
```

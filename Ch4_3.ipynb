```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/girl.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg1)
```

```python
hist = cv2.calcHist([gimg1], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
outimg = np.zeros(gimg1.shape, dtype='uint8')
```

```python
min = 0
max = 200

for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    if gimg1[h, w] < min:
      outimg[h, w] = 0
    elif gimg1[h, w] > max:
      outimg[h, w] = 255
    else:
      val = np.int32(gimg1[h, w] - min) * 255
      val = np.int32(val / (max - min))
      outimg[h, w] = val

cv2_imshow(outimg)
```

```python
hist2 = cv2.calcHist([outimg], [0], None, [256], [0, 256])
plt.plot(hist2)
```
